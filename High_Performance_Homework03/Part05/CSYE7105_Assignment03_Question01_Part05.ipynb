```python
from dask.distributed import Client
client = Client(n_workers=4)
client
```

```python
import dask                                                    # importing dask library
import dask.dataframe as dd
from nycflights13 import flights                               #from nycflights13 import flights
df = dd.from_pandas(flights, npartitions=4)                    #starting a Dask Client and set “n_workers=4” for this client.
df.head()
```

```python
print("Before cleaning the Nan values from the dataframe: {}".format(flights['dep_delay'].isnull().sum()))
```

```python
# Removing the samples with null values in the feature “dep_delay” in this dataframe
flights.dropna(subset=['dep_delay'], how ='all', inplace = True)
```

```python
print("After cleaning the NaN Values from the dataframe: {}".format(flights['dep_delay'].isnull().sum()))
```

Interpretation: This shows that the dataframe is clean and all the null values are being removed

```python
print(flights.info())
```

```python
# Using dask dataframe to compute the mean for departure delay “dep_delay” of all flights
dep_delay_mean = df['dep_delay'].mean().compute()
dep_delay_mean
```

```python
#Using dask dataframe to compute the standard deviation for departure delay “dep_delay” of all flights
dep_delay_std = df['dep_delay'].std().compute()
dep_delay_std
```
